# This Notebook Creates a Time-Series Model Using Prophet Framework and Makes a Prediction on our data for the Next 4 Weeks (1 Month). This is an example of a bounded CLV model

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)

summary_data = pd.read_csv('../../data/source/gt_omsa__org_usage_summary.csv')
reads_data = pd.read_csv('../../data/source/gt_omsa__daily_read_events.csv')
writes_data = pd.read_csv('../../data/source/gt_omsa__daily_write_events.csv')
```

```python
# Begin Reads EDA by measuring auto-correlation

def measure_autocorrelation(data, grain = 'reads'):
    
    '''Function to measure the auto-correlation in our datasets'''
    
    df = data.copy()
    
    if grain == 'reads':
    
        df['read_event_date'] = pd.to_datetime(df['read_event_date']) - pd.to_timedelta(7, unit='d')
        df = df.groupby(['org_id', pd.Grouper(key='read_event_date', freq='W-MON')])['sum_read_counter'].sum()        .reset_index()        .sort_values('read_event_date')

        df = df.groupby('read_event_date')['sum_read_counter'].sum().reset_index()
        plot_acf(x=df['sum_read_counter'], lags=50)
        plt.show()
        
    else:
        
        df['write_event_date'] = pd.to_datetime(df['write_event_date']) - pd.to_timedelta(7, unit='d')
        df = df.groupby(['org_id', pd.Grouper(key='write_event_date', freq='W-MON')])['sum_write_counter'].sum()        .reset_index()        .sort_values('write_event_date')

        df = df.groupby('write_event_date')['sum_write_counter'].sum().reset_index()
        plot_acf(x=df['sum_write_counter'], lags=50)
        plt.show()

measure_autocorrelation(reads_data, grain = 'reads')
```

```python
## Create a Weekly Time-series with our datasets in order to use Prophet to predict

def make_time_series(dataset, reads_usage_min = 8,writes_usage_min = 10 , grain = 'reads'):
    
    df = dataset.copy()
    
    #create logic to calculate training and testing dates 
    
    if grain == 'reads':
    
        df['read_event_date'] = pd.to_datetime(df['read_event_date']) - pd.to_timedelta(7, unit='d')
        df = df.groupby(['org_id', pd.Grouper(key='read_event_date', freq='W-MON')])['sum_read_counter'].sum()\
                        .reset_index()\
                        .sort_values('read_event_date')
        
        #remove last date - data is incomplete 
        
        max_tp = df['read_event_date'].max()
        df = df.loc[(df['read_event_date'] < max_tp)]
        
        max_data_date = df['read_event_date'].max()
        
        train_df = df[df['read_event_date'] <= max_data_date].reset_index(drop = True)
        
        anchor_date = train_df['read_event_date'].max() - timedelta(weeks = 7)
        
        train_df = train_df[train_df['read_event_date'] >= anchor_date].reset_index(drop = True)
        train_df = train_df[train_df['org_id'].map(train_df['org_id'].value_counts()) == reads_usage_min]\
                                    .reset_index(drop = True)
        
  
        testing_date = train_df['read_event_date'].max() - timedelta(weeks = 3)
        test_df = train_df[train_df['read_event_date']  >=  testing_date].reset_index(drop = True)
    
        train_df = train_df.loc[(train_df['read_event_date'] < testing_date)].reset_index(drop = True)
        
    
    else:
        
        df['write_event_date'] = pd.to_datetime(df['write_event_date']) - pd.to_timedelta(7, unit='d')
        df = df.groupby(['org_id', pd.Grouper(key='write_event_date', freq='W-MON')])['sum_write_counter'].sum()\
                        .reset_index()\
                        .sort_values('write_event_date')
        
                #remove last date - data is incomplete 
        
        max_tp = df['write_event_date'].max()
        df = df.loc[(df['write_event_date'] < max_tp)]
        
        max_data_date = df['write_event_date'].max()
        
        train_df = df[df['write_event_date'] <= max_data_date].reset_index(drop = True)
        
        anchor_date = train_df['write_event_date'].max() - timedelta(weeks = 9)
        
        train_df = train_df[train_df['write_event_date'] >= anchor_date].reset_index(drop = True)
        train_df = train_df[train_df['org_id'].map(train_df['org_id'].value_counts()) == writes_usage_min]\
                                    .reset_index(drop = True)
        
        testing_date = train_df['write_event_date'].max() - timedelta(weeks = 3)
        test_df = train_df[train_df['write_event_date']  >=  testing_date].reset_index(drop = True)
    
        train_df = train_df.loc[(train_df['write_event_date'] < testing_date)].reset_index(drop = True)
            
    
    return train_df , test_df


training_data, testing_data  = make_time_series(reads_data)
```

```python
#Tranform the dataset into a format usable by prophet

def create_prophet_object(dataset, grain = 'reads'):
    
    '''This function transform our dataset into a format acceptable by Prophet'''
    
    df = dataset.copy()
    
    if grain == 'reads':
    
        df = df.rename(columns = {'sum_read_counter':'y', 'read_event_date':'ds'})
        df = df[['ds','org_id','y']]
        df['ds'] = df['ds'].dt.tz_localize(None)
        
    else:
        
        df = df.rename(columns = {'sum_write_counter':'y', 'write_event_date':'ds'})
        df = df[['ds','org_id','y']]
        df['ds'] = df['ds'].dt.tz_localize(None)
        
    return df 

prophet_train = create_prophet_object(training_data)
prophet_test = create_prophet_object(testing_data)
```

```python
## Function to fit the model

def fit_prophet_object(group, periods = 4):
    
    '''This Function fits the prophet Model and makes a prediction for the next month (4 weeks) ''' 

    m = Prophet()
    m.fit(group)
  
    future = m.make_future_dataframe(periods=periods)
    forecast = m.predict(future)[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
    forecast['org_id'] = group['org_id'].iloc[0]
  
    # Return the forecasted results
    return forecast[['ds', 'org_id', 'yhat', 'yhat_upper', 'yhat_lower']] 
```

```python
## Function to make predictions on new data with our Fitted Prophet Model 

def prophet_predict(train):
    
    '''This Function makes predictions on our test set with our fitted Prophet Model'''

    start_time = time()
    for_loop_forecast = pd.DataFrame()
    
    for org in train['org_id'].unique().tolist():
        group = train[train['org_id'] == org] 
        forecast = fit_prophet_object(group)
        for_loop_forecast = pd.concat((for_loop_forecast, forecast))
        
    print('The time used for the for-loop forecast is ', time()-start_time)
    
    return for_loop_forecast
```

```python
# Make predictions with prophet model 

pred_data = prophet_predict(prophet_train)
```

```python
#Functions to evaluate our models with MAPE and WAPE

def mape(y_true, y_pred):
    ape = np.abs((y_true - y_pred)/ y_true)
    
    return np.mean(ape)

def wmape(y_true,y_pred):
    return np.sum(np.abs(y_true - y_pred) / np.sum(np.abs(y_true)))
```

```python
def evaluate_model(prediction_df, train,test):
    
    '''This function calculates the error between our predicted values and actual values'''
    
    max_date = train['ds'].max()
    
    preds = prediction_df.loc[(prediction_df['ds'] > max_date)]
    
    overall_df = pd.DataFrame()
    
    for i, v in list(zip(sorted(list(test['ds'].unique())),sorted(list(preds['ds'].unique())))):
        
        test_temp = test[test['ds'] == i]
        preds_temp = preds[preds['ds'] == v]
        test_temp = test_temp.merge(preds_temp, how = 'inner', on = 'org_id').rename(columns = {'ds_x':'ds'})
        test_temp['yhat'] = [i if i > 0 else 0 for i in test_temp['yhat']]
        test_temp = test_temp[['org_id','ds','yhat','y']]
        test_temp['error'] = abs(test_temp['y'] - test_temp['yhat'])
        overall_df = overall_df.append(test_temp)
        print('WAPE:', test_temp['error'].sum() / test_temp['y'].sum())
        
    return overall_df
        
reads_prediction_results = evaluate_model(pred_data, prophet_train, prophet_test)
```

```python
## Get the reads prediction error

def calculate_results(predictions, grain = 'reads'):
    
    '''Function to Calculate the actual CLV and the Predicted CLV using the unit-cost for reads and writes'''

    grouped_preds = predictions.groupby(['org_id','ds'])['y','yhat'].sum().reset_index()

    # calculate unit cost 
    
    if grain == 'reads':
        
        grouped_preds['total_org_cost'] =(grouped_preds['y']/1000000)* 0.33
        grouped_preds['predicted_org_cost'] = (grouped_preds['yhat']/1000000)* 0.33
        grouped_preds['prediction_error'] = abs(grouped_preds['total_org_cost'] - grouped_preds['predicted_org_cost'])
        grouped_preds['event_type'] = [''.join('read ').split() * len(grouped_preds['org_id'])][0]
        
    else:
        
        grouped_preds['total_org_cost'] =(grouped_preds['y']/1000000)* 0.49
        grouped_preds['predicted_org_cost'] = (grouped_preds['yhat']/1000000)* 0.49
        grouped_preds['prediction_error'] = abs(grouped_preds['total_org_cost'] - grouped_preds['predicted_org_cost'])
        grouped_preds['event_type'] = [''.join('write ').split() * len(grouped_preds['org_id'])][0]
            
    print('Prediction Error for Prophet Model: ', sum(grouped_preds['prediction_error'] / sum(grouped_preds['total_org_cost'])))
    
    return grouped_preds

reads_cost = calculate_results(reads_prediction_results)
```

```python
## This function gets the overall reads and writes CLV 

def get_overall_clv(dataset_reads,dataset_writes):
    
    '''Function to combine our reads and writes predicted CLV and compare them to actual CLV'''
    
    overall_df = dataset_reads.append(dataset_writes)
    overall_df = overall_df.groupby(['org_id'])['total_org_cost','predicted_org_cost'].sum().reset_index()
    
    print('overall model error:', wmape(overall_df['total_org_cost'],overall_df['predicted_org_cost']))
    print('overall model accuracy:', 1 - (wmape(overall_df['total_org_cost'],overall_df['predicted_org_cost'])))
    return overall_df
```

```python
#Calculate the writes CLV and estimate the error with WAPE 

writes_cost = calculate_results(writes_prediction_results, grain = 'writes')
```

```python
##Combine the reads and writes datasets and measure the overall CLV Error
# between the predicted CLV and the actual CLV

clv_data = get_overall_clv(reads_cost, writes_cost)
```
